```python
import pandas as pd
import re
import emoji
from tqdm import tqdm
from google.colab import files

# ğŸ“ íŒŒì¼ ì—…ë¡œë“œ
uploaded = files.upload()
file_name = list(uploaded.keys())[0]

# â³ tqdmê³¼ pandas ì—°ë™
tqdm.pandas()

# ğŸ“„ ì—‘ì…€ íŒŒì¼ ì½ê¸°
df = pd.read_excel(file_name)

# ğŸ§¹ ì´ëª¨ì§€ + ë¹„ì˜ì–´ ë¬¸ì ì œê±° í•¨ìˆ˜
def clean_text(text):
    if isinstance(text, float):
        text = str(text)
    text = emoji.replace_emoji(text, replace="")
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    return text.strip()

# ğŸ”„ ì „ì²˜ë¦¬ ì ìš© (ê¸°ì¡´ 'Review' ë®ì–´ì“°ê¸°)
df['Review'] = df['Review'].progress_apply(clean_text)

# âœ¨ ë¦¬ë·° ê¸¸ì´ ê³„ì‚° í•¨ìˆ˜
def get_review_length(text):
    try:
        return len(str(text).split())
    except:
        return 0

# ğŸ“ Review_length ì»¬ëŸ¼ ìƒì„±
df['Review_length'] = df['Review'].progress_apply(get_review_length)

# ğŸš« ë¦¬ë·° ê¸¸ì´ 10 ë¯¸ë§Œ ì œê±°, ê¸°ì¤€ì„ ë°”ê¾¸ë ¤ë©´ ì—¬ê¸°ì— ìˆëŠ” ìˆ«ì 10ì„ ìˆ˜ì •
df = df[df['Review_length'] >= 10].reset_index(drop=True)

# âœ… ì—‘ì…€ë¡œ ì €ì¥
preprocessed_file_path = 'preprocessed_comments_with_length.xlsx'
df.to_excel(preprocessed_file_path, index=False)
print(f"âœ… ì „ì²˜ë¦¬ + ë¦¬ë·° ê¸¸ì´ â‰¥10 í•„í„°ë§ ë°ì´í„°ë¥¼ '{preprocessed_file_path}' íŒŒì¼ë¡œ ì €ì¥í–ˆìŠµë‹ˆë‹¤.")

# ğŸ“¥ ë‹¤ìš´ë¡œë“œ
files.download(preprocessed_file_path)
```

#2. ëœë¤ Sampling

```python
import pandas as pd
from google.colab import files

# ì—‘ì…€ íŒŒì¼ ì—…ë¡œë“œ
uploaded = files.upload()
file_name = list(uploaded.keys())[0]

# ì—‘ì…€ íŒŒì¼ ë¶ˆëŸ¬ì˜¤ê¸°
df = pd.read_excel(file_name)

# ì „ì²´ ê°œìˆ˜ í™•ì¸
print(f"ì „ì²´ ë°ì´í„° ìˆ˜: {len(df)}ê°œ")

# ìƒ˜í”Œë§ ê°œìˆ˜ ì…ë ¥
n = int(input("ëª‡ ê°œë¥¼ ëœë¤ ìƒ˜í”Œë§í• ê¹Œìš”? ìˆ«ìë¥¼ ì…ë ¥í•˜ì„¸ìš”: "))

# ìœ íš¨ì„± ê²€ì‚¬
if n > len(df):
    raise ValueError("âŒ ìƒ˜í”Œ ìˆ˜ê°€ ì „ì²´ ë°ì´í„°ë³´ë‹¤ ë§ìŠµë‹ˆë‹¤.")

# ëœë¤ ìƒ˜í”Œë§
sampled_df = df.sample(n=n, random_state=42).reset_index(drop=True)

# ìƒ˜í”Œë§ ê²°ê³¼ ì €ì¥
sampled_file = f"random_sample_{n}.xlsx"
sampled_df.to_excel(sampled_file, index=False)

print(f"âœ… {n}ê°œì˜ ëœë¤ ìƒ˜í”Œ ë°ì´í„°ë¥¼ '{sampled_file}' íŒŒì¼ë¡œ ì €ì¥í–ˆìŠµë‹ˆë‹¤.")
files.download(sampled_file)
```
